```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Parameters
n = 10000   #number of data
N = 1000     #steps

#coin flip
c1 = np.random.choice([-1,1], size = (n,N))

#Gaussian distribution
mean = 0 
variance = 1
c2 = np.random.normal(mean, np.sqrt(variance), size = (n,N))

#Equal prob of being -1,(1-np.sqrt(3))/2,(1+np.sqrt(3))/2
c3 = np.random.choice([-1,(1-np.sqrt(3))/2,(1+np.sqrt(3))/2], size = (n,N))
mean_c3 = np.mean(c3)
variance_c3 = np.var(c3)

random_walks = [c1,c2,c3]
#plot
plt.figure(figsize=(15,5))

plt.subplot(1, 3, 1)
plt.hist(c1[0], bins=20, color ='r')
plt.xlabel("x")
plt.ylabel("p(x)")
plt.xlim((-4,4))
plt.title(f"Filp Coin")

plt.subplot(1, 3, 2)
plt.hist(c2[0], bins=20,color ='g')
plt.xlabel("x")
plt.ylabel("p(x)")
plt.xlim((-4,4))
plt.title(f"Gaussian distribution")

plt.subplot(1, 3, 3)
plt.hist(c3[0], bins=20,color ='b')
plt.xlabel("x")
plt.ylabel("p(x)")
plt.xlim((-4,4))
plt.title(f"-1,(1-np.sqrt(3))/2,(1+np.sqrt(3))/2")
print(f"Variance :  {variance_c3},  Mean : {mean_c3}")
```

```python
#excercise 5.1a
position_c1 = np.cumsum(c1,axis = 1)
position_c2 = np.cumsum(c2,axis = 1)
position_c3 = np.cumsum(c3,axis = 1)

#plot
plt.figure(figsize=(20,5))

plt.subplot(1, 3, 1)
plt.plot(position_c1[:100,:].T,linewidth=0.1,color = 'r')
plt.xlabel("step")
plt.ylabel("x(t)")
plt.ylim((-100,100))
plt.title(f"Filp Coin")

plt.subplot(1, 3, 2)
plt.plot(position_c2[:100,:].T,linewidth=0.1,color = 'g')
plt.xlabel("step")
plt.ylabel("x(t)")
plt.ylim((-100,100))
plt.title(f"Gaussian distribution")

plt.subplot(1, 3, 3)
plt.plot(position_c3[:100,:].T,linewidth=0.1,color = 'b')
plt.xlabel("step")
plt.ylabel("x(t)")
plt.ylim((-100,100))
plt.title(f"-1,(1-sqrt(3))/2,(1+sqrt(3))/2")
```
