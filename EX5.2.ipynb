```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def calculate_MSD(positions):
    N = len(positions)
    msd = np.zeros((1,N-1))

    for t in range(1, N):
        msd[:,t - 1] = np.mean((positions[t:] - positions[:-t])**2)
    return msd

```

```python
def fiveSecondTrajectories(n, dt):
    data = n
    num_step = int(5/dt)
    time = np.linspace(0, 5, num = num_step)
    x_t = np.zeros((data,num_step))
    for n  in range(data):
        for t  in range(1,num_step):
            w = np.random.randn()
            x_t[n,t] = x_t[n,t-1] + w*np.sqrt(dt)
    
    msd_trajectories = np.zeros((data,num_step - 1))
    for n in range(data):
        msd_trajectories[n] = calculate_MSD(x_t[n,:])
    msd = np.mean(msd_trajectories,axis=0)

    return x_t,time, msd
```

```python
#exercise 5.2a
n = 50      #number of trajectories

x_t_1,time_1,msd_1 = fiveSecondTrajectories(n,0.01)
x_t_2,time_2,msd_2 = fiveSecondTrajectories(n,0.05)
x_t_3,time_3,msd_3 = fiveSecondTrajectories(n,0.1)

#plot
plt.figure(figsize=(20,5))

plt.subplot(1, 3, 1)
plt.plot(time_1,x_t_1.T,linewidth=0.5,color = 'r')
plt.xlabel("t [s]")
plt.ylabel("x(t)")
plt.ylim(-8,8)
plt.title(f"dt = 0.01")

plt.subplot(1, 3, 2)
plt.plot(time_2,x_t_2.T,linewidth=0.5,color = 'g')
plt.xlabel("t [s]")
plt.ylabel("x(t)")
plt.ylim(-8,8)
plt.title(f"dt = 0.05")

plt.subplot(1, 3, 3)
plt.plot(time_3,x_t_3.T,linewidth=0.5,color = 'b')
plt.xlabel("t [s]")
plt.ylabel("x(t)")
plt.ylim(-8,8)
plt.title(f"dt = 0.1")
```
